```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=(r"C:\Users\praji\Downloads\house_price.csv")
df=pd.read_csv(data)
df
```

```python
#mean
df.price_per_sqft.mean()
```

```python
 #percentail method
df.price_per_sqft.quantile([0.25,0.50,0.75])
```

```python
q1=df.price_per_sqft.quantile(0.25)
```

```python
q3=df.price_per_sqft.quantile(0.75)
```

```python
q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
lower_whisker=q1-1.5*IQR
upper_whisker=q3+1.5*IQR
print ("Lower Wisker: ",lower_whisker)
print("Upper Wisker: ",upper_whisker)
```

```python
data_outlier=df[(df.price_per_sqft<lower_whisker)|(df.price_per_sqft>upper_whisker)]
data_outlier
```

```python
df['price_per_sqft'].describe().round()
```

```python
# Non Outlier
data=df[(df.price_per_sqft>lower_whisker) & (df.price_per_sqft<upper_whisker)]
data
```

```python
sns.histplot(data['price_per_sqft'], kde=True)
plt.xlabel('Price per Sqft')
plt.ylabel('Frequency')
plt.title('Distribution of Price per Sqft')
plt.show()
```

```python
m=df.price_per_sqft.mean()
s=df.price_per_sqft.std()
print (m)
print(s)
```

```python
df["Zscore"]=(df.price_per_sqft-m)/s
df
```

```python
df[(df.Zscore>-3) & (df.Zscore<3)]
```
